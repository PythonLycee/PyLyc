        <ul>
            <li>$a_1$ est la <strong>moyenne</strong> des longueurs $a_0$ et $b_0$ (dimensions du rectangle précédent);</li>
            <li>$b_1$ est choisi de telle sorte que l'aire du rectangle soit égale à $p$.</li>
        </ul>
    </li>
</ul>

## <span style="color:#8E44AD" id="2">2. Étude mathématique et algorithmique</span>

<strong>2.1. Justifier que $\forall n\in \mathbb{N}\;;\;\displaystyle a_{n+1}=\frac{a_n+b_n}{2}\;$ et $\;\forall n\in \mathbb{N}\;;\;\;\displaystyle b_{n}=\frac{p}{a_{n}}$.
<br><br>
2.2. En déduire que la suite $(a_n)_{n \geq 0}$ vérifie :
$ \begin{Bmatrix}
   a_0=p  \\
   \forall n\in \mathbb{N} \;;\; a_{n+1} = \displaystyle \frac{1}{2} \left( a_n+\frac{p}{a_n} \right) \end{Bmatrix}$
</strong>

<br><strong style="color:red">
2.1. $a_{n+1}$ étant par construction la moyenne de $a_n$ et $b_n$, on a $\displaystyle a_{n+1}=\frac{a_n+b_n}{2} \;$.<br>
$\quad\;\;$Pour tout $n\in\mathbb{N}$, l'aire du rectangle de rang $n$ a pour aire $a_n\times b_n = p$ donc $\displaystyle b_{n}=\frac{p}{a_{n}}$.
<br><br>
2.2. $a_0=p$ d'après l'énoncé (construction du rectangle de rang $0$).<br>$\displaystyle \quad\;\; a_{n+1}=\frac{a_n+b_n}{2}=\frac{a_n+\frac{p}{a_{n}}}{2}=\frac{1}{2}\left( a_n+\frac{p}{a_n} \right) \;$
</strong>


<strong>
2.3. Dans cette question, on suppose que $p=13$, de telle sorte que l'on a :<br><br>
$\quad\quad \begin{Bmatrix}
   a_0=13  \\
   \forall n\in \mathbb{N} \;;\; a_{n+1} = \displaystyle \frac{1}{2} \left( a_n+\frac{13}{a_n} \right) \end{Bmatrix}$       
    <br><br>
$\quad\;$On souhaite implémenter en Python la suite $(a_n)_{n\geq0}$.
<br><br>
$\quad\;\;$a. Écrire une fonction Python <mark style="font-family: Consolas;">Heron</mark> qui reçoit <mark style="font-family: Consolas;">n</mark> en argument et qui renvoie la valeur de $a_n$.
</strong>

```python
# Écrire ici la fonction Heron

def Heron(n):
    a = 13
    for k in range(n):
        a = 1/2*(a+13/a)
    return a
```

</strong>

```python
# Effectuer ici l'appel à la fonction Heron

Heron(3)
```

<strong>
2.4.a. Démontrer que $ \displaystyle \;\forall n\in \mathbb{N}\;; a_{n+1}^2-p = \frac{ \left( a_n^2-p \right)^2  }{4 a_n^2}$ et en déduire que $\;\forall n\in \mathbb{N} \;;\; a_n \geq \sqrt{p}$.
<br><br>
$\quad\;$b. Justifier que $\displaystyle \;\forall n\in \mathbb{N} \;;\; a_{n+1}-a_n = \frac{p-a_n^2}{2a_n}$ et en déduire que la suite $(a_n)_{n \geq 0}$ est décroissante.<br>
$\quad\;$c. En déduire que la suite $(a_n)_{n \geq 0}$ admet une limite finie, que l'on notera $L$.</strong>

<br><br>
<strong style="color:red">
2.4.a. $ \displaystyle \;\forall n\in \mathbb{N}\;; a_{n+1}^2-p = \left( \frac{1}{2}\left( a_n+\frac{p}{a_n} \right) \right) ^2 -p = \frac{1}{4} \left( \frac{a_n^2+p}{a_n} \right)^2 -p = \frac{ \left( a_n^2+p \right)^2 }{4a_n^2} -p = \frac{ \left( a_n^2+p \right)^2 -4pa_n^2}{4a_n^2} = \frac{ \left( a_n^2 \right)^2 +2pa_n^2+p²-4pa_n^2}{4a_n^2} = \frac{ \left( a_n^2 \right)^2 -2pa_n^2+p²}{4a_n^2} = \frac{ \left( a_n^2-p \right)^2  }{4 a_n^2}$<br>
Le membre de droite est un carré donc est positif, et on en déduit que $\forall n\in \mathbb{N}\;; a_{n+1}^2-p \geq 0$, ce qui donne $a_{n+1}\geq\sqrt{p}$.<br>
De plus, on a aussi $a_0=p=\sqrt{p}^2\geq\sqrt{p}$ car $p\in \mathbb{N}^*$.
<br><br>
b. 
$\displaystyle \;\forall n\in \mathbb{N} \;;\; a_{n+1}-a_n = \displaystyle \frac{1}{2} \left( a_n+\frac{p}{a_n} \right)-a_n =   \frac{1}{2} \left( a_n+\frac{p}{a_n}-2a_n \right) = \frac{1}{2} \left( \frac{p}{a_n}-a_n \right) = \frac{1}{2} \left( \frac{p-a_n^2}{a_n} \right)=\frac{p-a_n^2}{2a_n}$
<br>
$\;\forall n\in \mathbb{N} \;;\; a_n \geq \sqrt{p}$ donc $a_n^2 \geq p$ donc $p-a_n^2 \leq 0$, d'où on déduit $a_{n+1}-a_n \leq 0$, c'est à dire $a_{n+1}\leq a_n$.<br>
$(a_n)_{n \geq 0}$ est bien décroissante.
<br><br>
c. La suite $(a_n)_{n \geq 0}$ étant décroissante et minorée par $\sqrt{p}$, on en déduit qu'elle est convergente, et on note $L$ sa limite.
</strong>



<strong>
2.5. Justifier que la limite $L$ de la suite $(a_n)_{n \geq 0}$ vérifie $L = \displaystyle \frac{1}{2} \left( L+\frac{p}{L} \right)$. En déduire la valeur de $L$.
</strong>

<br><br>
<strong style="color:red">
2.5.<br> En passant à la limite quand $n$ tend vers $+\infty$ dans $a_{n+1} = \displaystyle \frac{1}{2} \left( a_n+\frac{p}{a_n} \right)$, on obtient $L = \displaystyle \frac{1}{2} \left( L+\frac{p}{L} \right)$
<br>
On résout alors l'équation :<br>
$L = \displaystyle \frac{1}{2} \left( L+\frac{p}{L} \right)$<br>
$\displaystyle \Leftrightarrow L - \frac{1}{2} \left( L+\frac{p}{L} \right) = 0$<br>
$\displaystyle \Leftrightarrow \frac{1}{2} L - \frac{p}{2L} = 0$<br>   
$\displaystyle \Leftrightarrow \frac{L^2-p}{2L} = 0$<br>
$\displaystyle \Leftrightarrow L^2-p = 0$<br>
$\displaystyle \Leftrightarrow (L-\sqrt{p})(L+\sqrt{p}) = 0$<br>
$\displaystyle \Leftrightarrow L \in \left\{ -\sqrt{p} ; \sqrt{p} \right\}$<br>
et comme $L \geq 0$ (limite d'une suite à termes positifs), on en déduit que $L = \sqrt{p}$.
</strong>


<strong>2.6.a. Modifier la fonction Python <mark style="font-family: Consolas;">Heron</mark> écrite à la question 2.3.a de telle sorte que :<br>
<ul>
    <li> la fonction reçoit deux arguments <mark style="font-family: Consolas;">p</mark> et <mark style="font-family: Consolas;">n</mark> ;</li>
    <li> la fonction renvoie la valeur de $a_n$ dans le cas général, où $a_0=p$.</li>
</ul>    
    
  

```python
# Écrire ici la nouvelle fonction Heron

def Heron(p,n):
    a = p
    for k in range(n):
        a = 1/2*(a+p/a)
    return a
```

```python
# Effectuer ici l'appel à la fonction Heron

Heron(5,4) # (on a calculé ici le terme de rang 4 de la suite)
```

## <span style="color:#8E44AD" id="3">3. Contexte historique</span>

<strong>Dans le tome 1 de son ouvrage "Metrica", Héron d'Alexandrie est amené à évaluer la racine carrée de 720.<br>Il écrit :</strong>

<BLOCKQUOTE style="background-color:Lavender;">
(...)<br>
έπεί ούν αί ψκ ρητήν τήν πλευράν ούκ έχουσι, ληψόμεθα μετά διαφόρου ελάχιστου τήν πλευράν ούτως.<br><br>
έπεί ό συνεγγίζων τώ ψκ τετράγωνός έστιν ό ψκθ καί πλευράν έ'χει τον κζ, μέρισον τάς ψκ εις τον κζ* γίγνεται κς καί τρίτα δύο’ πρόσθες τάς κζ* γίγνεται νγ τρίτα δύο* τούτων τό ήμισυ* γίγνεται κς L·' γ'* έσται άρα του ψκ ή πλευρά έγγιστα τά κς Δ.' γ'* τά γάρ κς ΙΔ γ' έφ’ έαυτά* γίγνεται ψκ λς'* ώστε τό διάφορον μονάδος έστί μόριον λς'<br><br>
εάν δέ βουλώμεθα έν έλάσσονι μορίω του λς' τήν διαφοράν γίγνεσθαι, άντί του ψκθ τάξομεν τά νϋν εύρεθέντα ψκ καί λς' καί ταύτά ποιήσαντες εύρήσομεν πολλω έλάττονα του λς' τήν διαφοράν γιγνομένην.
<br>(...)
</BLOCKQUOTE>
<strong>dont voici la traduction :</strong>
<BLOCKQUOTE style="background-color:Lavender;">
(...)<br>
Alors, puisque les 720 n’ont pas le côté exprimable, nous prendrons le côté avec une très petite différence ainsi.<br><br>
Puisque le carré le plus proche de 720 est 729 et qu’il a 27 comme côté, divise les 720 par le 27 : il en résulte 26 ⅔ ; ajoute les 27 : il en résulte 53 ⅔ ; de ceux-ci, la moitié : il en résulte 26 ½ ⅓ : le côté de 720 sera donc 26 ½ ⅓ à très peu près. En effet, les 26 ½ ⅓ par eux-mêmes : il en résulte 720 <span style="font-size:10px">$1/36$</span> ; de sorte que la différence est une 36e part d’unité.<br><br>
Et si nous voulons que la différence se produise par une part plus petite que 1/36, au lieu de 729, nous placerons les 720 <span style="font-size:10px">$1/36$</span> maintenant trouvés et, en faisant les mêmes choses, nous trouverons qu’il en résulte une différence inférieure, de beaucoup, à 1/36.
<br>
(...)
<br><br>
<span style="font-size:11px; color:#8E44AD">
NB : 
<ul>
    <li>La notation 26 ⅔ signifie $26+\frac{2}{3}$.</li>
    <li>La notation 26 ½ ⅓ signifie $26+\frac{1}{2}+\frac{1}{3}$.</li>
</ul>
    </span>    
</BLOCKQUOTE >



<strong>Ayant remarqué que le carré le plus proche de $720$ est $729=27^2$, Héron utilise comme valeur initiale $27$.<br>
En langage mathématique moderne, pour approcher $\sqrt{720}$, Héron utilise donc la suite $(u_n)_{n\geq0}$ définie par :<br><br>
$ \begin{Bmatrix}
   u_0=27  \\
   \forall n\in \mathbb{N} \;;\; u_{n+1} = \displaystyle \frac{1}{2} \left( u_n+\frac{720}{u_n} \right) \end{Bmatrix}$

</strong>

<strong>3.1. Vérifier par un calcul fractionnaire exact que la valeur trouvée par Héron pour $u_1$ est correcte. 
</strong>

<br><br>
<strong style="color:red">
3.1.<br>$u_1 = \displaystyle \frac{1}{2} \left( u_0+\frac{720}{u_0} \right) = \frac{1}{2} \left(27+\frac{720}{27} \right) = \frac{1}{2} \left(27+\frac{80}{3} \right) = \frac{1}{2} \times \frac{27\times3+80}{3} = \frac{161}{6}$.
<br>
Or on a $\displaystyle 26+\frac{1}{2}+\frac{1}{3}=\frac{26\times6+3+2}{6}=\frac{161}{6}$, ce qui prouve bien que  $\displaystyle u_1=26+\frac{1}{2}+\frac{1}{3}$.
</strong>

<strong>
3.2. Calculer $u_1^2$ et vérifier que l'erreur commise en remplaçant $720$ par $u_1^2$ est bien $\displaystyle \frac{1}{36}$. </strong>
<br>
<br>
<strong style="color:red">
3.2.<br>$\displaystyle u_1^2=\left( \frac{161}{6} \right)^2 = \frac{161^2}{6^2}=\frac{25921}{36}$<br>
$720\times36=25920$ donc $\displaystyle u_1^2=\frac{25920}{36}+\frac{1}{36}=720+\frac{1}{36}$
</strong>

## <span style="color:#8E44AD" id="4">4. Encadrement d'une racine carrée</span>

<strong>On reprend la suite $(a_n)_{n\geq 0}$ définie dans <a href="#2">la partie 2</a>, vérifiant : $ \begin{Bmatrix}
   a_0=p  \\
   \forall n\in \mathbb{N} \;;\; a_{n+1} = \displaystyle \frac{1}{2} \left( a_n+\frac{p}{a_n} \right) \end{Bmatrix}$ ,<br>
ainsi que la suite $(b_n)_{n\geq 0}$ vérifiant : $\displaystyle \forall n\in \mathbb{N} \;;\; b_n=\frac{p}{a_n}$.

<BLOCKQUOTE style="float:right; font-size: 11px; background-color:Lavender;">
        <strong>Info :</strong><br><br>
        Les deux suites $(a_n)_{n\geq0}$ et $(b_n)_{n\geq0}$ étudiées ici sont dites <strong>adjacentes</strong>.<br>
        <ul>
            <li>$(a_n)_{n\geq0}$ et $(b_n)_{n\geq0}$ sont respectivement décroissante et croissante ;</li>
            <li>$(a_n)_{n\geq0}$ et $(b_n)_{n\geq0}$ ont la même limite $\sqrt{p}$.</li>
        </ul>
        Il en résulte que $b_n\leq \sqrt{p} \leq a_n$ fournit un encadrement de $\sqrt{p}$ d'amplitude $a_n-b_n$.<br>     
        <img src="https://raw.githubusercontent.com/PythonLycee/PyLyc/master/img/Héron_adjacent.png">
</BLOCKQUOTE>    


<strong>4.1.a. On rappelle que la suite $(a_n)_{n \geq 0}$ est décroissante.<br>
$\quad\quad$Justifier que $(b_n)_{n \geq 0}$ est croissante.<br><br>
$\quad\;$b. On rappelle que la suite $(a_n)_{n \geq 0}$ a pour limite $\sqrt{p}$.<br>
$\quad\quad$En déduire la limite de la suite $(b_n)_{n \geq 0}$.
</strong>

<strong style="color:red">
4.1.a.<br>
$\displaystyle \forall n\in \mathbb{N} \;;\; b_{n+1}-b_n = \frac{p}{a_n+1} - \frac{p}{a_n} = \frac{p(a_n-a_{n+1})}{a_n a_{n+1}}$ et comme $a_{n+1} \leq a_n$, on en déduit que $a_n-a_{n+1}\geq0$ et donc $b_{n+1}-b_n\geq0$.<br>
La suite $(b_n)_{n\geq 0}$ est bien croissante. 
<br><br>
4.1.b.<br>
$\displaystyle\lim\limits_{n \to +\infty}{a_n}=\sqrt{p}\;$ donc $\;\displaystyle\lim\limits_{n \to +\infty}{\frac{p}{a_n}}=\frac{p}{\sqrt{p}}=\sqrt{p}\;$ c'est à dire $\;\displaystyle\lim\limits_{n \to +\infty}{b_n}=\sqrt{p}$.
</strong>

<strong>4.2.a. Reprendre et modifier la fonction Python <mark style="font-family: Consolas;">Heron</mark> de la question 2.6.a pour qu'elle renvoie les deux valeurs $a_n$ et $b_n$.

```python
# Écrire ici la nouvelle fonction Heron

def Heron(p,n):
    a = p
    for k in range(n):
        a = 1/2*(a+p/a)
        b = p/a
    return a,b
```

<strong>4.3. Écrire une fonction Python <mark style="font-family: Consolas;">Encadrement_racine</mark> qui :
<ul style="list-style-type:circle">
    <li>reçoit en arguments la valeur <mark style="font-family: Consolas;">p</mark> et une valeur <mark style="font-family: Consolas;">h</mark> strictement positive ;</li>
    <li>renvoie deux valeurs $a_N$ et $b_N$ telles que $a_N-b_N<h$.</li>
</ul>
</strong>

```python
# Écrire ici la fonction Encadrement_racine

def Encadrement_racine(p,h):
    a = p
    b = 1
    while a-b>=h:
        a = 1/2*(a+p/a)
        b = p/a
    return a,b
```

<strong>$\quad$b. À l'aide d'un appel à la fonction <mark style="font-family: Consolas;">Encadrement_racine</mark>, obtenir un encadrement de $\sqrt{720}$ d'amplitude inférieure à $10^{-5}$.</strong>

```python
# Exécuter ici l'appel à la fonction Encadrement_racine

Encadrement_racine(720,10**-5)
```
