La fonction <mark style="font-family: Consolas; background-color:#E1C9FC">esperance</mark> fournie ci-dessous permet de calculer l'espérance d'une variable aléatoire à partir de listes <mark style="font-family: Consolas; background-color:#E1C9FC">x</mark> et <mark style="font-family: Consolas; background-color:#E1C9FC">p</mark> contenant respectivement les valeurs prises par la variable aléatoire et leurs probabilités associées.<br><br>
Exécuter les deux cellules Python pour obtenir l'espérance de la variable aléatoire définie précédemment.<br>
Donner une interprétation concrète du résultat obtenu.
</strong>

```python
# Exécuter cette cellule pour mettre la fonction esperance en mémoire

def esperance(x,p):
    "fonction qui renvoie l'espérance d'une variable aléatoire"
    E=0
    for k in range(len(x)):
        E = E + x[k]*p[k]
    return E
    
```

```python
# Exécuter cette cellule pour effectuer un appel à la fonction esperance
esperance(x,p)
```
