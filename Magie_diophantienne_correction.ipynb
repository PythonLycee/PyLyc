## <span style="color:#016BF7" id="1">1. Présentation du tour</span> 

<strong style="color:#3183ED">Situation étudiée :</strong><br>
<BLOCKQUOTE style='background-color:#CEE2FC;'>
<p>Lors d’une représentation, un magicien mentaliste annonce aux spectateurs :<br><br> 
« Prenez le numéro de votre jour de naissance et multipliez-le par 12. Prenez le rang de votre mois de naissance et multipliez-le par 31. Ajoutez les deux nombres obtenus et donnez moi le résultat. Je pourrai alors vous donner la date de votre anniversaire ».</p>
</BLOCKQUOTE>


<strong>Un spectateur annonce $328$ et en quelques secondes, le magicien déclare : "Votre anniversaire tombe le 17 avril !".
<br>
<ul style="list-style-type:None">
<li>1.1. Vérifier que pour une personne née le 17 avril, le programme de calcul donne effectivement le nombre $328$.</li><br>
<li>1.2. Écrire une fonction Python <mark>Calcul</mark> qui :
<ul style="list-style-type:circle">
    <li>reçoit en arguments deux valeurs entières <mark>j</mark> et <mark>m</mark> correspondant respectivement au jour et au mois de naissance d'un spectateur ;</li>
    <li>renvoie le nombre obtenu avec le programme de calcul décrit précédemment.</li>
</ul>
</li>
</strong>
      
<strong style="color:red">
1.1. Le calcul à effectuer est : $7\times12+4\times31=328$.    
</strong>

```python
# Écrire ici la fonction Calcul

def Calcul(j,m):
    return j*12+m*31
```

```python
# Effectuer ici l'appel à la fonction Calcul

Calcul(17,4)
```

</strong>

<strong style="color:red">
2.1.<br>
Si on note et respectivement $x$ et $y$ le numéro du jour de naissance et le rang du mois de naissance, le programme de calcul effectué par la magicien donne $12x+31y=242$, qui est donc l'équation diophantienne à résoudre.<br>

<br>
On en déduit que $PGCD(31;12)=1$ et que $-5\times31+13\times12=1$.<br>
En multipliant par $242$ on obtient : $-1210\times31+3146\times12=242$.<br>
Ainsi, le couple $(x_0;y_0)=(3146;-1210)$ est une solution particulière de l'équation $(E):12x+31y=242$.<br>
<br>
2.3.<br>
Supposons que $(x;y)$ soit une solution de $(E)$.<br>
Alors :<br>
$12x+31y=3146\times12-1210\times31$<br>
$12(x-3146)=31\times(-y-1210)$<br>
$31$ divise $12(x-3146)$ et $31$ est premier avec $12$, donc d'après le théorème de Gauss, $31$ divise $x-3146$.<br>
Ainsi $\exists k \in \mathbb{Z}$ tel que $x-3146=31k$, c'est à dire $x=3146+31k$.<br>
On en déduit alors $12\times31k=31\times(-y-1210)$, qui donne $12k=-y-1210$, c'est à dire $y=-1210-12k$.<br>
On a donc $(x;y)=(3146+31k;-1210-12k)$.<br>
<br>
Réciproquement, soit $(x;y)=(3146+31k;-1210-12k)$.<br>
On a alors : $12x+31y=12\times(3146+31k)+31\times(-1210-12k)=12\times3146-31\times1210=242$.<br>
<br>
L'ensemble des solutions de $(E)$ est donc:<br>
$S=\left\{ (3146+31k;-1210-12k) \; / \; k\in\mathbb{Z} \right\}$<br>
<br>
2.4.<br>
La condition $1\leq y \leq 12$ donne $1\leq -1210-12k \leq 12$, c'est à dire $1211\leq -12k \leq 1222$.<br>
En tenant compte du fait que $k$ est entier, on obtient $-101 \leq k \leq -101$, c'est à dire $k=-101$.<br>
La solution cherchée est donc unique, et on a :<br>
$(x;y)=(3146+31\times(-101);-1210-12\times(-101))=(15;2)$.<br>
Le spectateur est donc né un 15 février.<br>

</strong>

```python
# Écrire ici la fonction TestE

def TestE(x,y,c):
    return 12*x+31*y==c
```

<strong><ul style="list-style-type:None"><li>
<p>$\quad\;\;$b. Écrire une fonction Python <mark>Resolution</mark> qui :
<ul style="list-style-type:circle">    
    <li>reçoit en argument une valeur entière <mark>c</mark> ;</li>
    <li>teste l'un après les autres les couples d'entiers $(x,y)$ tels que $1\leq x \leq 31$ et $1\leq y \leq 12$ dans l'équation $12x+31y=c$ et renvoie le premier couple <mark>x,y</mark> qui convient.</li>
</ul>
    </p></li></strong>
$\quad\quad\;\;\;$Aide : Pour générer les entiers $k$ vérifiant $a\leq k \leq b$, on peut utiliser la syntaxe <mark>for k in range(a,b+1):</mark>.

```python
# Écrire ici la fonction Resolution

def Resolution(c):
    for x in range(1,32):
        for y in range(1,13):
            if TestE(x,y,c):
                return (x,y)
    return None
```

    </p></li></strong>

```python
# Effectuer ici l'appel à la fonction Resolution

Resolution(242)
```
