<strong>3. La fonction <mark style="font-family: Consolas; background-color:#E1C9FC">esperance</mark> fournie ci-dessous permet de calculer l'espérance d'une variable aléatoire à partir de listes <mark style="font-family: Consolas; background-color:#E1C9FC">x</mark> et <mark style="font-family: Consolas; background-color:#E1C9FC">p</mark> contenant respectivement les valeurs prises par la variable aléatoire et leurs probabilités associées.<br><br>
Exécuter les deux cellules Python pour obtenir l'espérance de la variable aléatoire définie précédemment.<br>
Donner une interprétation concrète du résultat obtenu.
</strong>

```python
# Exécuter cette cellule pour mettre la fonction esperance en mémoire

def esperance(x,p):
    "fonction qui renvoie l'espérance d'une variable aléatoire"
    E=0
    for k in range(len(x)):
        E = E + x[k]*p[k]
    return E
    
```

```python
# Exécuter cette cellule pour effectuer un appel à la fonction esperance
esperance(x,p)
```

<strong>4. a. Écrire une fonction <mark style="font-family: Consolas; background-color:#E1C9FC">variance</mark> qui permet de calculer la variance d'une variable aléatoire à partir de listes <mark style="font-family: Consolas; background-color:#E1C9FC">x</mark> et <mark style="font-family: Consolas; background-color:#E1C9FC">p</mark> contenant respectivement les valeurs prises par la variable aléatoire et leurs probabilités associées.
</strong><br>
Aide : On pourra s'inspirer de l'écriture de la fonction <mark style="font-family: Consolas; background-color:#E1C9FC">esperance</mark> donnée précédemment.


```python
# Écrire ici la fonction variance

def variance(x,p):
    "fonction qui renvoie la variance d'une variable aléatoire"
    V=0
    for k in range(len(x)):
        V = V + (x[k]-esperance(x,p))**2*p[k]
    return V

#OU

def variance(x,p):
    "fonction qui renvoie la variance d'une variable aléatoire"
    E=esperance(x,p) # Ceci permet de ne calculer qu'une seule fois l'espérance
    V=0
    for k in range(len(x)):
        V = V + (x[k]-E)**2*p[k]
    return V   
    
```

<strong>$\;\;$ b. Effectuer un appel à la fonction <mark style="font-family: Consolas; background-color:#E1C9FC">variance</mark> pour calculer la variance de la variable aléatoire précédemment définie.
</strong>

```python
# Effectuer ici l'appel à la fonction variance
variance(x,p) 
```

<strong>5. a. Écrire une fonction <mark style="font-family: Consolas; background-color:#E1C9FC">ecart_type</mark> qui permet de calculer l'écart-type d'une variable aléatoire à partir de listes <mark style="font-family: Consolas; background-color:#E1C9FC">x</mark> et <mark style="font-family: Consolas; background-color:#E1C9FC">p</mark> contenant respectivement les valeurs prises par la variable aléatoire et leurs probabilités associées.
</strong><br>
Aide : La syntaxe <mark style="font-family: Consolas; background-color:#E1C9FC">from math import sqrt</mark> permet d'utiliser la fonction Python <mark style="font-family: Consolas; background-color:#E1C9FC">sqrt</mark> pour le calcul d'une racine carrée.

```python
# Écrire ici la fonction ecart_type
from math import sqrt

def ecart_type(x,p):
    "fonction qui renvoie l'écart type d'une variable aléatoire"
    return sqrt(variance(x,p))
```

</strong>

```python
# Effectuer ici l'appel à la fonction ecart_type
ecart_type(x,p)
```
