Dans cette activité, on étudie une phase d'accélération d'un véhicule.<br><br>
Le protocole permet de mesurer la <strong style="color:#286E18">distance parcourue</strong> par le véhicule en fonction du <strong style="color:#3535F9">temps écoulé</strong>, pendant les 13 premières secondes d'accélération.<br>

<ul>

<BLOCKQUOTE style="background-color:#F1E8EB;">
<strong style="color:#AD1D55;">À savoir :</strong>
<br> 
    
Lorsque $\color{#6D4AB5}t$ est "proche" de $\color{#006699}a$, la droite $\color{#FE812A}{\text{(AM)}}$ se rapproche d'une droite $T$ appelée <strong>tangente à la courbe $C_f$ au point d'abscisse $\color{#006699}a$</strong>.<br><br>
Le coefficient directeur de cette droite $T$ s'appelle le <strong>nombre dérivé de $f$ en $a$</strong> et il se note <strong>$f'(a)$</strong>.<br><br>
    
Pour résumer la méthode de construction de $f'(a)$ vue précédemment, on note :<br>

<img src="https://raw.githubusercontent.com/PythonLycee/PyLyc/master/img/véhicule_vitesse_2.png" style="float:right; max-width:70%;">     
    
<BLOCKQUOTE style="background-color:#FFFFFF;">  
$\displaystyle f'(a)= \lim_{t\rightarrow a}{ \frac{\color{#286E18}{f(t)}-\color{#286E18}{f(a)}}{\color{#6D4AB5}{t}-\color{#006699}a}}$  
</BLOCKQUOTE>
    <br><br><br><br><br>
</BLOCKQUOTE>


<strong>
2.1. Reprendre la figure dynamique précédente et régler $\color{#006699}a=\color{#006699}{10}$.<br>
$\quad\;$ Par lecture graphique, estimer $f'(10)$, qui est la vitesse instantanée du véhicule à la $\color{#006699}{10}$<sup style="color:#006699">ème</sup> seconde de l'expérience.<br>
</strong>
$\quad\;$ On pourra approcher la valeur de $\color{#6D4AB5}t$ de celle de $\color{#006699}a$ et observer la droite $\color{#FE812A}{\text{(AM)}}$.
<br><br>
<strong>
2.2. Démontrer que $\displaystyle \frac{\color{#286E18}{f(t)}-\color{#286E18}{f(10)}}{\color{#6D4AB5}{t}-\color{#006699}{10}}=\color{#6D4AB5}{t}+\color{#006699}{10}$.<br>
$\quad\;$ Justifier que cette égalité permet de retrouver la valeur $f'(10)$.
<br><br>
2.3. Démontrer que $\displaystyle \frac{\color{#286E18}{f(t)}-\color{#286E18}{f(a)}}{\color{#6D4AB5}{t}-\color{#006699}{a}}=\color{#6D4AB5}{t}+\color{#006699}{a}$.<br>
$\quad\;$ En déduire la valeur de $f'(a)$, exprimée en fonction de $a$.
</strong>


<br><br><br>
<strong style="color:red;">
2.1. Pour $t$ "proche" de $a$, on peut lire graphiquement le coefficient directeur de la droite (AM).<br>
$\quad$ Par lecture graphique, on lit ainsi $\displaystyle f'(10)=\frac{100}{5}=20$. <br>
$\quad$ La vitesse du véhicule à la $10$<sup>ème</sup> seconde est donc de $20\; ms^{-1}$.
<br><br>
2.2. $\displaystyle \frac{f(t)-f(10)}{t-10}= \frac{t²-10²}{t-10}= \frac{(t-10)(t+10)}{t-10}= t+10$<br>
$\quad$ On en déduit donc (en considérant $t$ "proche" de $10$) que :<br>
$\quad$ $\displaystyle f'(10)= \lim_{t\rightarrow 10}{ \frac{f(t)-f(10)}{t-10} }= \lim_{t\rightarrow 10}{(t+10)} = 10+10=20$     <br><br> 
2.3. $\displaystyle \frac{f(t)-f(a)}{t-a}= \frac{t²-a²}{t-a}= \frac{(t-a)(t+a)}{t-a}= t+a$<br>
$\quad$ On en déduit donc (en considérant $t$ "proche" de $a$) que :<br>
$\quad$ $\displaystyle f'(a)= \lim_{t\rightarrow a}{ \frac{f(t)-f(a)}{t-a} }= \lim_{t\rightarrow 10}{(t+a)} = a+a=2a$  

</strong>
