## <span style="color:#0934E6" id="1">1. Résolution dans $\mathbb{R}$ d'une équation du second degré à coefficients réels</span>
<br>

On rappelle le résultat de cours suivant :<br>

<BLOCKQUOTE style="background-color:#D7DDF7">
        <strong style="color:#0934E6; font-size:10px">RÉSOLUTION DANS $\mathbb{R}$ D'UNE ÉQUATION DU SECOND DEGRÉ À COEFFICIENTS RÉELS</strong>
    <br><br>Soit $f(x)=ax^2+bc+c$ (avec $a\in\mathbb{R}^*$ ; $b\in\mathbb{R}$ ; $c\in\mathbb{R}$ ) une fonction polynôme du second degré.<br><br>
    On pose $\Delta = b^2-4ac$ , appelé discriminant de $f$.<br><br>
    <ul>
        <li>Si $\Delta > 0$ alors l'équation $f(x)=0$ a exactement <strong>deux solutions réelles</strong> :<br>
            <center>$\displaystyle x_1=\frac{-b-\sqrt{\Delta}}{2a} \;$ et $\; \displaystyle x_2=\frac{-b+\sqrt{\Delta}}{2a}$<br><br></center></li>
        <li>Si $\Delta = 0$ alors l'équation $f(x)=0$ a <strong>une unique solution réelle</strong> :<br>
            <center>$\displaystyle x_0=-\frac{b}{2a} \;$</center></li>
        <li>Si $\Delta < 0$ alors l'équation $f(x)=0$ n'a <strong>aucune solution réelle</strong>.</li>   
    </ul>
</BLOCKQUOTE>


<strong>1. Définir une fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">Delta</mark> qui reçoit en arguments les valeurs $a$ ; $b$ et $c$ et qui renvoie la valeur du discriminant $\Delta$ de $f$.<br>
$\quad$Effectuer ensuite un test avec $a=5$ ; $b=3$ et $c=-4$.</strong>

```python
# Écrire ici la fonction Delta

def Delta(a,b,c):
    return b**2-4*a*c
```

```python
# Effectuer ici un test du calcul de Delta pour a=5; b=3 et c=-4

Delta(4,8,-60)
```

<strong>2. On fournit ci-dessous une fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">Racines</mark>.<br>

</ol>   
    
<br><br>
<strong style="color:red">
2.b. Une erreur apparaît lorsque le discriminant $\Delta$ est non nul. En effet, dans ce cas la variable <mark style="color:red">sol</mark> n'a pas été initialisée, et Python renvoie une erreur pour <mark style="color:red">return sol</mark> car cette variable n'existe pas.<br>
2.c. Dans le cas où $\Delta=0$, la fonction renvoie une liste qui contient la liste des solutions de l'équation du second degré.
</strong>

```python
# Exécuter cette cellule

def Racines(a,b,c):

    d = Delta(a,b,c)
    
    if d==0:
        sol = [-b/(2*a)]
    
    return sol
    
```

<strong>3.a. Modifier la fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">Racines</mark> pour qu'elle fournisse dans tous les cas <u>la liste</u> des solutions réelles de l'équation du second degré $ax²+bx+c=0$.</strong>
<br><br>
Aides :
<ul>
    <li>La syntaxe <mark style="font-family: Consolas; background-color:#E9ECF9">from math import sqrt</mark> permet d'utiliser la fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">sqrt</mark> pour le calcul d'une racine carrée.</li>
    <li>La syntaxe <mark style="font-family: Consolas; background-color:#E9ECF9">[ ]</mark> permet de générer une liste vide.</li>
    <li>Pour une liste contenant plusieurs éléments, on sépare les éléments de la liste à l'aide de virgules.</li>
</ul>

```python
from math import sqrt

# Écrire ici la fonction Racines modifiée puis exécuter la cellule

def Racines(a,b,c):

    d = Delta(a,b,c)
    
    if d==0:
        sol = [-b/(2*a)]
    elif d>0:
        sol = [(-b-sqrt(d))/(2*a),(-b+sqrt(d))/(2*a)]
    else:
        sol = []
    
    return sol
```

On rappelle le résultat de cours suivant :<br>

<BLOCKQUOTE style="background-color:#D7DDF7">
    <strong style="color:#0934E6; font-size:10px">RÉSOLUTION DANS $\mathbb{C}$ D'UNE ÉQUATION DU SECOND DEGRÉ À COEFFICIENTS RÉELS</strong>
    <br><br>Soit $f(z)=az^2+bz+c$ (avec $a\in\mathbb{R}^*$ ; $b\in\mathbb{R}$ ; $c\in\mathbb{R}$ ) une fonction polynôme du second degré.<br><br>
    On pose $\Delta = b^2-4ac$ , appelé discriminant de $f$.<br><br>
    <ul>
        <li>Si $\Delta > 0$ alors l'équation $f(z)=0$ a exactement <strong>deux solutions, qui sont réelles</strong> :<br>
            <center>$\displaystyle x_1=\frac{-b-\sqrt{\Delta}}{2a} \;$ et $\; \displaystyle x_2=\frac{-b+\sqrt{\Delta}}{2a}$<br><br></center></li>
        <li>Si $\Delta = 0$ alors l'équation $f(z)=0$ a <strong>une unique solution, qui est réelle</strong> :<br>
            <center>$\displaystyle x_0=-\frac{b}{2a} \;$</center></li>
        <li>Si $\Delta < 0$ alors l'équation $f(z)=0$ a exactement <strong>deux racines complexes conjuguées</strong> :
            <center>$\displaystyle z_1=\frac{-b-i\sqrt{\vert \Delta \vert}}{2a} \;$ et $\; \displaystyle z_2=\overline{z_1}=\frac{-b+i\sqrt{\vert \Delta \vert}}{2a}$<br><br></center></li>   
    </ul>       
</BLOCKQUOTE>


<strong>2.1. Modifier la fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">Racines</mark> (obtenue à la question 3.a) pour qu'elle fournisse dans tous les cas <u>la liste</u> des solutions <u>complexes</u> d'une équation du second degré à coefficients réels.</strong>
<br><br>
Aides :
<ul>
    <li>La syntaxe <mark style="font-family: Consolas; background-color:#E9ECF9">from math import sqrt</mark> permet d'utiliser la fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">sqrt</mark> pour le calcul d'une racine carrée.</li>
    <li>La fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">abs</mark> permet de calculer une valeur absolue.</li>
    <li>La syntaxe <mark style="font-family: Consolas; background-color:#E9ECF9">[ ]</mark> permet de générer une liste vide.</li>
    <li>Pour une liste contenant plusieurs éléments, on sépare les éléments de la liste à l'aide de virgules.</li>
</ul>

```python
from math import sqrt

# Écrire ici la fonction Racines modifiée puis exécuter la cellule

def Racines(a,b,c):

    d = Delta(a,b,c)
    
    if d==0:
        sol = [ complex(-b/(2*a),0) ]
    elif d>0:
        sol = [ complex( (-b-sqrt(d))/(2*a),0 ) , complex( (-b+sqrt(d))/(2*a),0 )]
    else:
        sol = [ complex( -b/(2*a),-sqrt(abs(d))/(2*a) ) , complex( -b/(2*a),sqrt(abs(d))/(2*a) )   ]
    
    return sol
```

2.2.a. Voici un résumé dans chaque cas :

## <span style="color:#0934E6" id="3">3. Résolution dans $\mathbb{C}$ d'une équation du 3<sup>ème</sup> degré à coefficients réels </span><br><span style="color:#7D90DE;font-size:14px">$\quad\;$(Niveau Terminale Math Expertes)</span>

<strong>3.1. Dans cette question, on souhaite résoudre dans $\mathbb{C}$ l'équation du 3<sup>ème</sup> degré $g(z)=0$ avec :<br><br><center>$g(z)=z^3-13z^2+57z-85$</center></strong> 

<strong>$\quad\;\;$a. Définir une fonction Python <mark style="font-family: Consolas; background-color:#E9ECF9">g</mark> qui reçoit une valeur $x$ en argument et qui renvoie son image $g(x)$ par la fonction $g$.</strong> 

```python
# Écrire ici la fonction Python g

def g(z):
    return z**3-13*z**2+57*z-85
```

<strong>$\quad\;\;$b. Exécuter la cellule Python ci-dessous.<br>$\quad\quad\;\;$Que représente la valeur obtenue ? dans quel ensemble de nombres est-elle cherchée ?<br>$\quad\quad\;\;$Cette valeur sera notée $x_0$ dans les questions suivantes.</strong>

```python
# Exécuter cette cellule

def recherche_sol_elem(fonction):
    for k in range(-10,11):
        if fonction(k)==0 :
            return k
    return None
    
recherche_sol_elem(g)
```

<strong style="color:red">La valeur obtenue (ici $5$) est une solution particulière de l'équation $g(x)=0$, cherchée parmi les valeurs entières comprises entre $-10$ et $10$</strong>

<strong>$\quad\;\;$c. On admet le résultat suivant :</strong>
    
<BLOCKQUOTE style="background-color:#D7DDF7">
    Soit $g(x)$ une fonction polynomiale non nulle à coefficients réels de degré $n$.<br><br>
    Si $x_0$ est une racine réelle de $g$ (c'est à dire telle que $g(x_0)=0$), alors $g$ peut s'écrire :<br>
    $g(z)=(z-x_0)f(z)$ où $f$ est une fonction polynôme à coefficients réels de degré $n-1$.
</BLOCKQUOTE>

<strong>$\quad\quad\;\;$On pose $g(z)=(z-x_0)(az^2+bz+c)$, où $x_0$ est la valeur obtenue dans la question 3.1.b.<br>
$\quad\quad\;\;$Déterminer la valeur des coefficients $a$ ; $b$ et $c$.<br><br></strong>
$\quad\quad\;\;$Aide : On pourra développer l'expression $(z-x_0)(az^2+bz+c)$ et identifier les coefficients avec ceux de $z^3-13z^2+57z-85$.


<strong style="color:red">On développe :<br>
$g(z)=(z-x_0)(az^2+bz+c)$<br>
$\quad\;\;=(z-5)(az^2+bz+c)$<br>
$\quad\;\;=az^3+bz^2+cz-5az^2-5bz-5c$<br>
$\quad\;\;=az^3+(b-5a)z^2+(c-5b)z-5c$<br>
En identifiant les coefficients des deux écritures :<br>
$g(z)=az^3+(b-5a)z^2+(c-5b)z-5c=z^3-13z^2+57z-85$ ,<br>
on obtient :<br>
$\begin{Bmatrix} a=1 \\ b-5a=-13 \\ c-5b=57 \\ -5c=85 \end{Bmatrix}$<br>
ce qui donne $\begin{Bmatrix} a=1 \\ b=-8 \\ c=17 \end{Bmatrix}$<br>
On en déduit donc que :<br>
$g(z)=(z-5)(z^2-8z+17)$
</strong>


<strong>$\quad\;\;$d. Effectuer un appel à la fonction <mark style="font-family: Consolas; background-color:#E9ECF9">Racines</mark> obtenue dans la question 2.1, et en déduire l'ensemble des solutions complexes de l'équation $g(z)=0$. </strong>

```python
# Effectuer ici l'appel à la fonction Racines

Racines(1,-8,17)
```

<strong style="color:red">Les solutions complexes de $g$ sont $5$ ; $4-i$ et $4+i$.
