<br><br>
Aide : En langage Python, les virgules des nombres décimaux s'écrivent avec des points, et on sépare les nombres par des virgules.</li>
</ul>

```python
# abscisses du mobile (exprimées en cm)
x = [ 10.3 , 15.3 , 20.5 , 25.5 , 30.3 , 35.3 , 40.2 , 45.1 , 49.9 , 54.8 , 59.7 , 64.3 , 69.2 , 73.9 , 78.4 ] 

# ordonnées du mobile (exprimées en cm)
y = [ 40.4 , 38.8 , 37.3 , 35.7 , 34.4 , 32.8 , 31.3 , 29.8 , 28.4 , 26.9 , 25.5 , 24.0 , 22.5 , 21.1 , 19.7 ] 

# intervalle de temps entre deux prises de vue (exprimé en s)
Dt = 0.1
```

<BLOCKQUOTE style="background-color:#DFE6FF; color:#031E7F;">
<ul>
    <li><strong>Question 1.5. </strong><br>
Exécuter la cellule suivante, qui calcule les distances $\text{M}_i\text{M}_{i+1}$.<br>
À partir des résultats obtenus, indiquer quelle caractéristique du mouvement du mobile est mise en évidence.
    </li></ul>
</BLOCKQUOTE>


```python
from math import sqrt

def distance(x,y):
    "fonction qui calcule les distances à partir des listes d'abscisses et d'ordonnées"
    return [ round ( sqrt( (x[i+1]-x[i])**2 + (y[i+1]-y[i])**2 ) ,1) for i in range(len(x)-1) ]

distance(x,y)
```

<BLOCKQUOTE style="background-color:#DFE6FF; color:#031E7F;">
<ul>
    <li><strong>Question 1.6. </strong><br>Pour chaque point $M_i(x_i;y_i)$, on fait l'estimation que les coordonnées du vecteur vitesse $\vec{v_i}$ vérifient :<br>
<font size="3"> $\displaystyle \left( \vec{v_i} \right)_x = \frac{x_{i+1}-x_i}{Δt}$ </font> $\;$ et $\;$
<font size="3"> $\displaystyle \left( \vec{v_i} \right)_y = \frac{y_{i+1}-y_i}{Δt}$ </font>.
<br><br>
On a défini ci-dessous une liste Python <mark style="font-family: Consolas; background-color:#CAD4FC;">vx</mark> contenant les abscisses de ces vecteurs.<br>
Définir de la même façon la liste Python <mark style="font-family: Consolas; background-color:#CAD4FC;">vy</mark> contenant leurs ordonnées.<br><br>Note : la syntaxe Python <mark style="font-family: Consolas; background-color:#CAD4FC;">round( ,1)</mark> permet d'obtenir des valeurs approchées au dixième près.
</li>
</ul>
</BLOCKQUOTE>


```python
# liste des abscisses des vecteurs vitesse
vx = [ round ( (x[i+1]-x[i])/Dt ,1 ) for i in range(len(x)-1) ]

# liste des ordonnées des vecteurs vitesse (à créer)
vy = [ round ( (y[i+1]-y[i])/Dt ,1 ) for i in range(len(y)-1) ]

# affichage des listes
vx,vy
```

<BLOCKQUOTE style="background-color:#DFE6FF; color:#031E7F;">
<ul>
    <li><strong>Question 1.7. </strong><br>La fonction Python <mark style="font-family: Consolas; background-color:#CAD4FC;">Graph_Pointage</mark> ci-dessous permet d'obtenir le tracé des <span style="color: #0000FF">vecteurs vitesse</span> pour le <span style="color: #FF0000">pointage</span> effectué.<br>Exécuter cette cellule et la suivante pour obtenir le graphique.</li></ul>
</BLOCKQUOTE>


```python
#Exécuter cette cellule
import matplotlib.pyplot as plt

def Graph_Pointage(x,y,vx,vy,zoom_v=20,nb_fig=1):
    "Fonction représentant le pointage avec affichage des vecteurs vitesses"
     
    #initialisation du graphique et génération des légendes
    plt.figure(nb_fig)
    plt.clf()
    plt.title('Trajectoire du mobile')
    plt.xlabel("abscisse en cm",fontsize=8)
    plt.ylabel("ordonnée en cm",fontsize=8)    
    plt.arrow(0,0,10,0,head_width=2, head_length=0.5, fc='lightblue', ec='blue')
    plt.text(0,1,"unité de\nvitesses\n$"+str(10*zoom_v)+"\;cm\cdot s^{-1}$",fontsize=8)    
    
    # représentation des points
    plt.scatter(x,y,color='red') 
    
    # représentation des vecteurs vitesse
    for i in range(len(vx)):
        plt.arrow( x[i] , y[i] , vx[i]/zoom_v , vy[i]/zoom_v , head_width=0.5, head_length=1, fc='lightblue', ec='blue')
    
    # affichage
    plt.show()
    
```

```python
Graph_Pointage(x,y,vx,vy)
```
