<h1 style="color:#048D1B; font-size:40px">Diversité des immunoglobulines <span style="color:red">(Corrigé)</span></h1>
<br><br>




<BLOCKQUOTE style="background-color:#EDFCF0;">
   
<figure style="float: right;">
        <img src="https://raw.githubusercontent.com/PythonLycee/PyLyc/master/img/Immunoglobuline_schema.gif">
</figure>
Les molécules d’anticorps (immunoglobulines) sont constituées de quatre chaînes d’acides aminés : deux chaînes lourdes ou H (pour « heavy ») et deux chaînes légères ou L (pour « light »), les séquences peptidiques des chaînes lourdes étant identiques entre elles, et de même pour les deux chaînes légères.<br><br>
Des mécanismes génétiques particuliers sont à l’origine des gènes codant les chaînes H et L, et ainsi de la très grande diversité des anticorps produits par le système immunitaire. L’objectif de cette activité est de traiter l’un de ces mécanismes, de manière volontairement simplifiée, les processus mis en jeu étant complexes et multiples. Le mécanisme étudié correspond à la constitution d’un gène à partir de l’association de segments d’ADN issus d’un ensemble de segments différents disponibles dans le génome.<br><br>
L’activité permettra de dénombrer les différentes combinaisons de segments possibles dans un gène codant pour une chaîne d’anticorps.
  
    
</BLOCKQUOTE>

<h3 style="color:#048D1B; font-size:28px">3. Génération des chaînes lourdes</h3>



<BLOCKQUOTE style="background-color:#EDFCF0;">
Le mécanisme à l'origine du gène codant la chaîne lourde est similaire à celui qui vient d'être vu pour la chaîne légère.<br>
Les nombres de segments impliqués étant :
<ul>
    <li>40 exemplaires différents de segments $\text{V}_\text{H}$ ;</li>
    <li>27 exemplaires différents de segments $\text{D}_\text{H}$ ;</li>
    <li>6 exemplaires différents de segments $\text{J}_\text{H}$ ;</li>

<strong>
3.1. Écrire une instruction permettant de générer une structure <mark style="font-family: Consolas; background-color:#CBFCD3">liste_VH</mark> contenant les 40 segments $\text{V}_\text{H}$ possibles.<br>
</strong>
$\;\;\;$ Aide: On pourra commencer par effectuer un copier/coller de la structure fournie dans la question 2.1.


```python
# Exécuter cette cellule
liste_VH = [ "VH"+str(k) for k in range(1,40+1) ]
liste_VH
```

<strong>
3.2. Écrire une instruction permettant de générer une structure <mark style="font-family: Consolas; background-color:#CBFCD3">liste_DH</mark> contenant les 27 segments $\text{D}_\text{H}$ possibles.<br>
</strong>

```python
# Effectuer ici la saisie pour générer liste_JL
liste_DH = [ "DH"+str(k) for k in range(1,27+1) ]
liste_DH
```

<strong>
3.3. Écrire une instruction permettant de générer une structure <mark style="font-family: Consolas; background-color:#CBFCD3">liste_JH</mark> contenant les 6 segments $\text{J}_\text{H}$ possibles.<br>
</strong>

```python
# Effectuer ici la saisie pour générer liste_JL
liste_JH = [ "JH"+str(k) for k in range(1,6+1) ]
liste_JH
```

<strong>3.4. Exécuter la cellule ci-dessous, qui permet de générer une structure <mark style="font-family: Consolas; background-color:#CBFCD3">liste_chaines_lourdes</mark> représentant la liste des chaînes lourdes que l'on peut obtenir par combinaison.</strong>
<br><br>


```python
# Exécuter cette cellule
def combinaison_H():
    liste=[]
    for VH in liste_VH:
        for DH in liste_DH:
            for JH in liste_JH:
                liste.append( VH+"-"+DH+"-"+JH+"-CH")
    return liste

liste_chaines_lourdes = combinaison_H()
liste_chaines_lourdes 
```

<h3 style="color:#048D1B; font-size:28px">4. Assemblage des chaînes légères et lourdes</h3>

<strong>
4.1.a. Exécuter la cellule ci-dessous, qui permet de générer une structure <mark style="font-family: Consolas; background-color:#CBFCD3">liste_immunoglobulines</mark> contenant l'ensemble des immunoglobulines que l'on peut obtenir par assemblage.</strong>

```python
def assemblage():
    liste=[]
    for chaineL in liste_chaines_legeres:
        for chaineH in liste_chaines_lourdes:
            liste.append(chaineL+"-/-"+chaineH)
    return liste

liste_immunoglobulines = assemblage()
```

<strong>Quelques compléments :</strong><br>
<ul>
    <li>Comme on vient de le voir, la formation d’un gène fonctionnel codant pour une chaîne de la molécule d’anticorps implique la recombinaison de segments d’ADN et explique ainsi en partie la diversité des anticorps produits par le système immunitaire ; les possibilités sont encore multipliées par le fait que la molécule d’anticorps résulte de l’association de deux types de chaînes générées indépendamment l’une de l’autre.<br>
À cela s’ajoutent d’autres mécanismes génétiques particuliers, modifiant encore les séquences nucléotidiques de l’ADN codant pour les chaînes. Par exemple l’association entre le segment $\text{V}_\text{L}$ et le segment $\text{J}_\text{L}$ (ou entre le segment $\text{D}_\text{H}$ et le segment $\text{J}_\text{H}$) ne se fait pas systématiquement de manière précise : lors du réarrangement, des nucléotides peuvent être ôtés à l’extrémité du segment $\text{V}_\text{L}$. À l’inverse un autre phénomène peut aboutir à l’ajout de quelques nucléotides au segment $\text{D}_\text{H}$. Par ailleurs, des mutations surviennent également et modifient là encore la séquence nucléotidique de l’ADN codant les chaînes d’anticorps.</li><br>
    <li>Comme les mécanismes génétiques particuliers formant les gènes codant les chaînes $\text{H}$ et $\text{L}$ se produisent de manière aléatoire, certains aboutissent à former des gènes codant pour des anticorps se liant aux propres molécules de l’organisme : dans l’organisme existent des systèmes permettant d’inactiver et éliminer ces lymphocytes « auto-immunes ».</li>
</ul>
