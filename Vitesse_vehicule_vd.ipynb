<h1 style="color:#840F3C; font-size:40px">Vitesse instantanée d'un véhicule</h1>
<h2 style="color:#AD1D55; font-size:18px">Approche de la notion de nombre dérivé d'une fonction en un point <br></h2>


<BLOCKQUOTE style="background-color:#F1E8EB;">


   
    
Dans cette activité, on étudie une phase d'accélération d'un véhicule.<br><br>
Le protocole permet de mesurer la <strong style="color:#286E18">distance parcourue</strong> par le véhicule en fonction du <strong style="color:#3535F9">temps écoulé</strong>, pendant les 13 premières secondes d'accélération.<br>

<img src="https://raw.githubusercontent.com/PythonLycee/PyLyc/master/img/véhicule_vitesse.png" style="float:right; max-width:40%;">
    
<br>
On notera :
<ul>
    <li>$\color{#3535F9}t$ le temps écoulé exprimé en seconde ;</li>
    <li>$\color{#286E18}{f(t)}$ la distance parcourue exprimée en mètre.</li>
</ul>
<br>
On admettra que pour tout $t\in[0;13]$, on a : 
    
<BLOCKQUOTE style="background-color:#FFFFFF;">    
$\quad f(t)=t²$
</BLOCKQUOTE>
    
On note $C_f$ la courbe représentative de la fonction $f$ dans un repère orthogonal, et $\;\color{#FF0000}{\text{M}}(\color{#3535F9}{t}\;;\color{#286E18}{f(t)})$ un point de cette courbe.
</BLOCKQUOTE>

<BLOCKQUOTE style="background-color:#F1E8EB;">
<strong style="color:#AD1D55;">À savoir :</strong>
<br><br> 
    
Lorsque $\color{#6D4AB5}t$ est "proche" de $\color{#006699}a$, la droite $\color{#FE812A}{\text{(AM)}}$ se rapproche d'une droite $T$ appelée <strong>tangente à la courbe $C_f$ au point d'abscisse $\color{#006699}a$</strong>.<br><br>
Le coefficient directeur de cette droite $T$ s'appelle le <strong>nombre dérivé de $f$ en $a$</strong> et il se note <strong>$f'(a)$</strong>.<br><br>
    
Pour résumer la méthode de construction de $f'(a)$ vue précédemment, on note :<br>

<img src="https://raw.githubusercontent.com/PythonLycee/PyLyc/master/img/véhicule_vitesse_2.png" style="float:right; max-width:70%;">     
    
<BLOCKQUOTE style="background-color:#FFFFFF;">  
$\displaystyle f'(a)= \lim_{t\rightarrow a}{ \frac{\color{#286E18}{f(t)}-\color{#286E18}{f(a)}}{\color{#6D4AB5}{t}-\color{#006699}a}}$  
</BLOCKQUOTE>
    <br><br><br><br><br><br>
</BLOCKQUOTE>
