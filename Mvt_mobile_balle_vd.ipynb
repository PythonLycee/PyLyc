<strong style="color:red">Carole, je suis parti du principe qu'on utilise les coordonnées des vecteurs pour déterminer que le mvt est rectiligne uniforme. Mais s'il faut utiliser d'autres outils ou méthodes, dis moi... <br><br>
</strong>

```python
def coord_vect_succ(x,y):
    return [ ( x[k+1]-x[k] , y[k+1]-y[k] ) for k in range(len(x)-1) ]   

coord_vect_succ(x,y)
```
