```python
# abscisses du mobile (exprimées en cm)
x = [ 10.3 , 15.3 , 20.5 , 25.5 , 30.3 , 35.3 , 40.2 , 45.1 , 49.9 , 54.8 , 59.7 , 64.3 , 69.2 , 73.9 , 78.4 ] 

# ordonnées du mobile (exprimées en cm)
y = [ 40.4 , 38.8 , 37.3 , 35.7 , 34.4 , 32.8 , 31.3 , 29.8 , 28.4 , 26.9 , 25.5 , 24.0 , 22.5 , 21.1 , 19.7 ] 

# intervalle de temps entre deux prises de vue (exprimé en s)
Dt = 0.1
```

<BLOCKQUOTE style="background-color:#DFE6FF; color:#031E7F;">
<ul>
    <li><strong>Question 1.5. </strong><br>Pour chaque point $M_i(x_i;y_i)$, on fait l'estimation que les coordonnées du vecteur vitesse $\vec{v_i}$ vérifient :<br>
<font size="3"> $\displaystyle \left( \vec{v_i} \right)_x = \frac{x_{i+1}-x_i}{Δt}$ </font> $\;$ et $\;$
<font size="3"> $\displaystyle \left( \vec{v_i} \right)_y = \frac{y_{i+1}-y_i}{Δt}$ </font>.
<br><br>
On a défini ci-dessous une liste Python <mark style="font-family: Consolas; background-color:#CAD4FC;">vx</mark> contenant les abscisses de ces vecteurs.<br>
Définir de la même façon la liste Python <mark style="font-family: Consolas; background-color:#CAD4FC;">vy</mark> contenant leurs ordonnées.<br><br>
</li>
</ul>
</BLOCKQUOTE>
<strong style="color:red">Carole : ci-dessous, on ne donnerait que l'écriture de vx, l'élève devrait écrire vy par analogie. Je pourrais comme tu le suggérais ajouter une instruction pour arrondir les résultats... mais finalement puisqu'on fait une interprétation graphique ensuite à partir de laquelle on fait les observations, je pense que ce n'est pas nécessaire.Qu'en penses-tu ?<br>
    Pour info, la syntaxe serait <code>vx = [ round( (x[i+1]-x[i])/Dt ,2) for i in range(len(x)-1) ]</code>
<br><br>Pour les questions et le graphique, je me suis inspiré de cette activité : <a href="https://notebook.basthon.fr/?from=https://raw.githubusercontent.com/PythonLycee/PyLyc/master/Base_Ball_correction_vd.ipynb">https://notebook.basthon.fr/?from=https://raw.githubusercontent.com/PythonLycee/PyLyc/master/Base_Ball_correction_vd.ipynb</a> </strong>

```python
# liste des abscisses des vecteurs vitesse
vx = [ (x[i+1]-x[i])/Dt for i in range(len(x)-1) ]

# liste des ordonnées des vecteurs vitesse (à créer)
vy = [ (y[i+1]-y[i])/Dt for i in range(len(y)-1) ]

# affichage des listes
vx,vy
```

<BLOCKQUOTE style="background-color:#DFE6FF; color:#031E7F;">
<ul>
    <li><strong>Question 1.6. </strong><br>La fonction Python <mark style="font-family: Consolas; background-color:#CAD4FC;">Graph_Pointage</mark> ci-dessous permet d'obtenir le tracé des <span style="color: #0000FF">vecteurs vitesse</span> pour le <span style="color: #FF0000">pointage</span> effectué.<br>Exécuter cette cellule et la suivante pour obtenir le graphique.</li></ul>
</BLOCKQUOTE>
<strong style="color:red">Carole : Je suis obligé de changer l'unité pour les vecteurs, sinon c'est illisible. Du coup, j'ajoute une légende pour les vecteurs vitesse sur la figure (en bas à gauche)... qu'en penses-tu ? </strong>

```python
#Exécuter cette cellule
import matplotlib.pyplot as plt

def Graph_Pointage(zoom_v=20):
    "Fonction représentant le pointage avec affichage des vecteurs vitesses"
    
    #initialisation du graphique et génération des légendes
    plt.clf()
    plt.title('Trajectoire du mobile')
    plt.xlabel('abscisses en cm',fontsize=8)
    plt.ylabel('ordonnées en cm',fontsize=8)    
    plt.arrow(0,0,10,0,head_width=0.5, head_length=1, fc='lightblue', ec='blue')
    plt.text(0,1,"unité de\nvitesses\n$"+str(10*zoom_v)+"\;cm\cdot s^{-1}$",fontsize=8)    
    
    # représentation des points
    plt.scatter(x,y,color='red') 
    
    # représentation des vecteurs vitesse
    for i in range(len(vx)):
        plt.arrow( x[i] , y[i] , vx[i]/zoom_v , vy[i]/zoom_v , head_width=0.5, head_length=1, fc='lightblue', ec='blue')
    
    # affichage
    plt.show()
```

```python
Graph_Pointage()
```
